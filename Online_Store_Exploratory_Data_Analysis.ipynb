### Робота із пропущеними наченнями

У таблиці products пропущені значення відсутні. Кількість пустих значень у таблиці countries є незначною, тому видалимо їх.

Колонка Country Code таблиці events містить 6% пропущених значень. Ця колонка є ключовою для об'єднання таблиць countries та events. Оскільки способів відновити цю інформацію не було знайдено, рядки із пропущеними значеннями доведеться видалити.

```python
# видаляємо рядки з пропущеними значеннями
countries_clean = countries.dropna()
events_clean = events.dropna()
```

```python
# перевіряємо
print(countries_clean.info())
print(events_clean.info())
```

```python
# переводимо колонки з датами в адекватний формат
events_clean["Order Date"] = pd.to_datetime(events_clean["Order Date"])
events_clean["Ship Date"] = pd.to_datetime(events_clean["Ship Date"])
```

```python
# перевіряємо
events_clean.info()
```

```python
# перевіряємо на наявність дублікатів
print(events_clean.duplicated().sum())
print(countries_clean.duplicated().sum())
print(products.duplicated().sum())
```

```python
# об'єднуємо таблиці
events_countries = pd.merge(events_clean, countries_clean, left_on="Country Code", right_on="alpha-3", how="left")
data = pd.merge(events_countries, products, left_on="Product ID", right_on="id", how="left")
data.head() # перевіряємо
```

```python
data.columns = data.columns.str.lower().str.replace(" ", "_").str.replace("-", "_") # перейменовуємо назви колонок у snake_case
data = data.drop(["order_priority", "country_code", "product_id", "alpha_2", "alpha_3", "id"], axis=1) # видаляємо зайві колонки
data.head() # перевіряємо
```

```python
data["sales_channel"] = data["sales_channel"].str.lower() # перетворюємо значення колонки sales_channel в єдиний формат
data["sales_channel"].value_counts() # перевіряємо
```

# Аналіз та візуалізація даних

```python
# створюємо колонки для аналізу тоталів у найпершій таблиціі (до очистки даних)
events["total_revenue"] = events["Units Sold"] * events["Unit Price"]
events["total_cost"] = events["Units Sold"] * events["Unit Cost"]
events["total_profit"] = events["total_revenue"] - events["total_cost"]

print("Загальні показники неочищених даних:\n")
print(f"Загальна кількість замовлень: {events['Order ID'].nunique()}")
print(f"Загальний дохід: {events['total_revenue'].sum()}")
print(f"Загальна собівартість: {events['total_cost'].sum().round(2)}")
print(f"Загальна прибуток: {events['total_profit'].sum()}")
```

```python
# створюємо аналогічні колонки уже в очищених даних
data["total_revenue"] = data["units_sold"] * data["unit_price"]
data["total_cost"] = data["units_sold"] * data["unit_cost"]
data["total_profit"] = data["total_revenue"] - data["total_cost"]

print("Загальні показники очищених даних:\n")
print(f"Загальна кількість замовлень: {data['order_id'].nunique()}")
print(f"Загальний дохід: {data['total_revenue'].sum()}")
print(f"Загальна собівартість: {data['total_cost'].sum().round(2)}")
print(f"Загальна прибуток: {data['total_profit'].sum()}")
print(f"Загальна кількість охоплених країн: {data['name'].nunique()}")
```

Оскільки нам довелося видалити 6% даних, для аналізу загальних показників беремо оригінальну таблицю з повним набором даних. Таким чином ми уникнемо неточностей у тоталах. У комірці вище можна побачити тотали очищених даних.

Бачимо загальний розподіл кількості днів, витрачених на доставку. Розподіл схожий на рівномірний. Найчастішою була доставка в районі 50 днів. Проте кількість таких доставок значним чином не відрізняється від інших кількостей.
